```python
import matplotlib.pyplot as plt
import requests
import pandas as pd
from collections import defaultdict
from bs4 import BeautifulSoup as bs
import csv
```

```python
url='https://en.wikipedia.org/wiki/List_of_countries_by_past_and_future_population'
res=requests.get(url).text
```

```python
outfile=open("table_data.csv","w",newline='')
writer=csv.writer(outfile)
soup=bs(res,"html")
table=soup.find('table')
list_of_rows=[]
for row in table.find_all('tr'):
    list_of_cells=[]
    for cell in row.find_all(["th","td"]):
        text=cell.text
        list_of_cells.append(text)
    list_of_rows.append(list_of_cells)
for item in list_of_rows:
    writer.writerow(item)
    #print (' '.join(item))
    
```

```python
df=pd.read_csv('table_data.csv')
df.head()
```

```python
plotcountries=['Austria','Germany','United States','France']
for country in plotcountries:
    row=df.iloc[country]
    plt.plot(row.index,row,label=row.name)
plt.ylim(ymin=0)
plt.xticks(rotation=70)
plt.legend(loc='best')
plt.xlabel("Year")
plt.ylabel("# people(million)")
plt.title("Population of countries")
```
